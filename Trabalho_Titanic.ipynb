2. Calcule a probabilidade de sobrevivência com base em diferentes características, como gênero ou classe, e compare-as.

```python
import matplotlib.pyplot as plt

# Carregando os dados de treinamento
train_data = pd.read_csv('train.csv')

# Probabilidade de sobrevivência por gênero
gender_prob = train_data.groupby('Sex')['Survived'].mean()

# Probabilidade de sobrevivência por classe socioeconômica
class_prob = train_data.groupby('Pclass')['Survived'].mean()

# Gráfico de barras para probabilidade de sobrevivência por gênero
plt.figure(figsize=(8, 6))
gender_prob.plot(kind='bar', color=['pink', 'blue'], alpha=0.7)
plt.title('Probabilidade de sobrevivência por gênero')
plt.xlabel('Gênero')
plt.ylabel('Probabilidade de sobrevivência')
plt.xticks(rotation='horizontal')
plt.show()

# Gráfico de barras para probabilidade de sobrevivência por classe socioeconômica
plt.figure(figsize=(8, 6))
class_prob.plot(kind='bar', color='green', alpha=0.7)
plt.title('Probabilidade de sobrevivência por classe socioeconômica')
plt.xlabel('Classe socioeconômica')
plt.ylabel('Probabilidade de sobrevivência')
plt.xticks(rotation='horizontal')
plt.show()
```

Probabilidade de sobrevivência por gênero:

Mulheres:   74.20%
Homens:     18.89%

------------------------------------------------
Probabilidade de sobrevivência por classe socioeconômica:

1:    62.96%

2:    47.28%

3:    24.23%

A probabilidade de sobrevivência ser significativamente maior para mulheres do que para homens no naufrágio do Titanic pode ser atribuída a diversos fatores:

1- Política de evacuação: Durante o naufrágio, a prioridade foi dada às mulheres e crianças para embarcar nos botes salva-vidas. A tripulação seguiu uma política de "mulheres e crianças primeiro", o que aumentou as chances de sobrevivência das mulheres.

2- Normas sociais e valores da época: No início do século XX, havia expectativas sociais de proteger as mulheres e garantir sua segurança. Essas normas influenciaram o comportamento de passageiros e tripulação no momento do resgate.

3- Localização dos passageiros: As cabines e áreas de alojamento das mulheres estavam localizadas em regiões mais próximas aos botes salva-vidas. Isso facilitou o acesso delas aos botes e aumentou suas chances de sobrevivência.

4- Resistência física e habilidades de natação: As mulheres podem ter tido maior resistência física para suportar as condições adversas, como a água fria, e podem ter tido maior habilidade de natação em comparação aos homens.
